```python
import pandas as pd
import spacy
df=pd.read_csv('/content/test[1].csv')
pln=spacy.load('en_core_web_sm')
print(df)
```

```python
d=df['text'].iloc[0]
print(d)
```

```python
doc=pln(d)
for i in doc.sents:
  print(i.text)
for i in doc:
  print(i.text)
```

```python
from nltk.stem import PorterStemmer
st=PorterStemmer()
for i in doc:
  print(st.stem(i.text))
for j in doc:
  print(st.stem(j.text))
```
